```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
#y(x) = k∙cos(x – a) + b
x = np.linspace(-2*np.pi, 3*np.pi, 201)
plt.plot(x, np.cos(x - 2))
plt.plot(x, np.cos(x - 2) + 0.5)
plt.plot(x, 2*np.cos(x - 2) + 0.5)
plt.xlabel('x')
plt.xlabel('y')
plt.grid(True)
plt.show()
```
