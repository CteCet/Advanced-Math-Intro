```python
import numpy as np
import itertools
```

```python
for p in itertools.permutations("12349", 5):
    print(''.join(str(x) for x in p))
```

```python
for p in itertools.combinations("012349", 5):
    print(''.join(p))
```

```python
for p in itertools.product("010", repeat=4):
    print(''.join(p))
```
