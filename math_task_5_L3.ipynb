```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#1
from pylab import*
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
Z = 2*X + 3*Y - 1
ax.plot_wireframe(X, Y, Z)
Z = 2*X + 3*Y + 12
ax.plot_wireframe(X, Y, Z)

show()
```

```python
#2
from pylab import*
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
Z = np.sqrt(X**2+Y**2)
ax.plot_surface(X, Y, Z)
show()
```
