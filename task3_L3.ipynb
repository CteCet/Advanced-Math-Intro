```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
#1
x = np.linspace(-1.5, 1.5, 200)
y = np.linspace(-1.5, 1.5, 200)
X, Y = np.meshgrid(x,y)
R = np.sqrt(X**2 + Y**2) 
fig, ax = plt.subplots()
ax.contour(X,Y,R,[1])
ax.set_aspect(1)
plt.grid(linestyle='--')
plt.show()
```

```python
#2
x = np.linspace(-1.5, 1.5, 200)
y = np.linspace(-1.0, 1.0, 200)
X, Y = np.meshgrid(x,y)
R = np.sqrt(X**2 + 4*Y**2)
fig, ax = plt.subplots()
ax.contour(X,Y,R,[1])
plt.grid(linestyle='--')
plt.show()
```

```python
#3
x = np.linspace(-1.5, 1.5, 200)
y = np.linspace(-1.0, 1.0, 200)
X, Y = np.meshgrid(x,y)
R = 25*X**2 - 4*Y**2
fig, ax = plt.subplots()
ax.contour(X,Y,R,[1])
plt.grid(linestyle='--')
plt.show()
```
