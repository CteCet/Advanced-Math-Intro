```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
#1
#y = x2 – 1
#exp(x) + x∙(1 – y) = 1
x = np.linspace(-10, 10, 201)
plt.plot(x, (np.exp(x) + x -1)/x)
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-2, 16)
plt.grid(True)
plt.show()
```

```python
from scipy.optimize import fsolve

def f(p):
    x, y = p
    return(y - x**2 + 1, np.exp(x) + x-x*y - 1)
    
x1, y1 = fsolve(f, (3, 8))
x1, y1 = fsolve(f, (-2, 2))
print(x1, y1)
```

```python
#2
#y = x2 – 1
#exp(x) + x∙(1 – y) > 1
from scipy.optimize import fsolve

def f(p):
    x, y = p
    return(y - x**2 + 1, (np.exp(x) + x-x*y - 1)>0)
    
x1, y1 = fsolve(f, (-1, 1.5))
print(x1, y1)
```
