```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-50, 150, 201)
z = np.linspace(-50, 150, 201)
plt.plot(x,z, 0.5 - 0.5*x - 0.5*z)
plt.plot(x, -1.75 + 0.75*x)
plt.plot(x,z, -2.4 + 1.6*x + 0.4*z)
plt.plot(x,z, 2*x - 5*z -7)
plt.plot(x,z, 3.75 - 2.75*x + z)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
A = np.array([[1, 2, -1], [3, -4, 0], [8, -5, 2], [2, 0,-5], [11, 4, -7]])
B = np.array([1, 7, 12, 7, 15])
np.linalg.lstsq(A,B)
```

```python
def Q(x, y, z):
    return((np.linalg.norm(np.dot(A, [x, y, z]) - B))**2)
Q(1.13919353, -0.90498444, -0.9009803)
```
