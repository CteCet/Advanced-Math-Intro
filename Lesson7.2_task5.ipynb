```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Q(x, y, z):
    return (x**2 + y**2 + z**2)

x = np.linspace(-10, 45, 100)
z = np.linspace(-10, 45, 100)
plt.plot(x,z, Q(x,z, 0.5 - 0.5*x + 0.5 *z))
plt.plot(x,z, Q(x,z, -2.4 + 1.6*x + 0.4 *z))
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
